```python
expenses = []

while True:
  print("\nEnter your expense: ")

  date = input('Data (YYYY-MM-DD): ')
  category = input('Category (e.g., Food, Transport): ')

  try:
    amount = float(input('Amount: '))
  except ValueError:
    print('Amount invalid! Try again')
    continue

  #Store each expense as a DICTIONARY
  expense = {
      'date': date,
      'category': category,
      'amount': amount
  }

  expenses.append(expense)

  #Ask if user wanna go on
  more = input('Add another? (yes/no): ').lower()
  if more != 'yes':
    break

#Summary
print("\n-------Summary-----")
print(f"Total entries: {len(expenses)}")

totalSpent = sum(e['amount'] for e in expenses) #I dont get this line of code, whats e'amount'?
print(f"Total spent: {totalSpent} USD")

#Print each entry nicely
print("\nYour Expenses: ")
for e in expenses:
  print(f"{e['date']} - {e['category'].title()} - {e['amount']} VND")
```
