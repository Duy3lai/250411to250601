```python
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 100, 1)

decay_fast = 0.1
decay_slow = 0.03

remaining_fast = np.exp(-decay_fast * time)
remaining_slow = np.exp(-decay_slow * time)

# Calculate half-life
half_life_fast = np.log(2) / decay_fast
half_life_slow = np.log(2) / decay_slow

print("Fast decay half-life:", half_life_fast)
print("Slow decay half-life:", half_life_slow)

plt.plot(time, remaining_fast, label='Fast Decay (0.1)', color='red')
plt.plot(time, remaining_slow, label='Slow Decay (0.03)', color='green')

plt.axvline(half_life_fast, color='red', linestyle='--', label='Fast Half-Life')
plt.axvline(half_life_slow, color='green', linestyle='--', label='Slow Half-Life')

plt.title("Comparison of Radioactive Decay with Half-Lives")
plt.xlabel("Time")
plt.ylabel("Remaining Substance")
plt.legend()
plt.grid(True)
plt.show()
```
