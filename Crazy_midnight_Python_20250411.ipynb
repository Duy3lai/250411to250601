```python
import numpy as np

person1 = np.array([60,165])
person2 = np.array([65,170])

diff = person2 - person1
distance = np.linalg.norm(diff)

print(f'Difference vector: {diff}')
print(f'Euclidean distance between person1 and person2: {distance}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
origin = np.array([[0, 0], [0, 0]])  # Start at origin
v1 = np.array([3, 1])
v2 = np.array([1, 2])

# Stack the vectors for plotting
vectors = np.array([v1, v2]).T

# Plot
plt.quiver(*origin, vectors[0], vectors[1], angles='xy', scale_units='xy', scale=1, color=['r', 'b'])
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid()
plt.gca().set_aspect('equal')
plt.title("Vector Plot")
plt.show()
```

```python
import numpy as np

np.random.seed(42)
scores = np.random.normal(loc=75, scale=10, size=100)

print('First 10 scores:', scores[:10])

#Analyze the fake data
print('Average score:', scores.mean())
print('Highest score:', scores.max())
print('Lowest score:',scores.min())
print('Standard deviation:', scores.std())

#Filter Students Who Scored Above 90
above90 = scores[scores > 90]
print('Students scoring above 90:', above90)
print('How many?', len(above90))

#Bonus: Add Another Feature (like Study Hours)
studyHours = np.random.normal(loc=5, scale=1.5, size=100)

#Basic correlation (how related the two are)
correlation = np.corrcoef(scores,studyHours)[0,1]
print('Correlation between scores and study hours:',correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Always reproducible
np.random.seed(1)

# Step 1: Fake study hours for 50 students
study_hours = np.random.normal(loc=2, scale=2, size=50)  # mean 5 hrs, std dev 2

# Step 2: Add noise to simulate real-life ups & downs
noise = np.random.normal(0, 8, 50)

# Step 3: Scores are based on hours studied + randomness
scores = 10 * study_hours + noise

# Step 4: Correlation
correlation = np.corrcoef(study_hours, scores)[0, 1]
print(f"Correlation: {correlation:.2f}")

# Step 5 (Optional): Visualize
plt.scatter(study_hours, scores)
plt.title("Study Hours vs Scores")
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
studyHours = np.random.normal(5,2,50)
noise = np.random.normal(0,5,50)
scores = studyHours * 10 + noise
scores = np.clip(scores,0,100)

#Linear regression (least squares)
coeffs = np.polyfit(studyHours, scores, 2) #slope, intercept

#Predict scores
predictedScores = np.polyval(coeffs, studyHours)

#plot
plt.scatter(studyHours, scores, label='Actual Data')
plt.plot(studyHours, predictedScores, color='red', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.title('Study Hours vs Score Prediction')
plt.legend()
plt.grid(True)
plt.show()

newPredictions = np.polyval(coeffs, newHours)
newHours = np.array([1,3,5,7])
for hours, predicted in zip(newHours, newPredictions):
  print(f'If a student studies {hours} hour(s), predicted score: {predicted:.2f}')
```
