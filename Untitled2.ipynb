```python
import matplotlib.pyplot as plt

# Bridge parameters
span = 10         # meters
height = 1.5      # truss height in meters
num_panels = 6    # number of truss segments (must be even for clean Warren)
incline_deg = 10  # bridge incline angle

import numpy as np
incline_rad = np.radians(incline_deg)

# Generate deck node positions
dx = span / num_panels
deck_nodes = [(i*dx, np.tan(incline_rad)*i*dx) for i in range(num_panels + 1)]

# Generate bottom chord (truss) nodes
bottom_nodes = [(x, y - height) for (x, y) in deck_nodes]

# Combine for plotting
nodes = deck_nodes + bottom_nodes
top_indices = list(range(len(deck_nodes)))
bot_indices = list(range(len(deck_nodes), len(deck_nodes)*2))

# Connectivity for Warren Truss (bottom)
members = []

# Top chord (deck)
for i in range(num_panels):
    members.append((top_indices[i], top_indices[i+1]))

# Bottom chord
for i in range(num_panels):
    members.append((bot_indices[i], bot_indices[i+1]))

# Diagonals
for i in range(num_panels):
    if i % 2 == 0:
        members.append((top_indices[i], bot_indices[i+1]))
    else:
        members.append((top_indices[i+1], bot_indices[i]))

# Verticals (optional, to reinforce deck)
for i in range(len(deck_nodes)):
    members.append((top_indices[i], bot_indices[i]))

# Plotting
fig, ax = plt.subplots(figsize=(12, 4))
for (i, j) in members:
    x_vals = [nodes[i][0], nodes[j][0]]
    y_vals = [nodes[i][1], nodes[j][1]]
    ax.plot(x_vals, y_vals, 'k-')

# Mark nodes
for x, y in nodes:
    ax.plot(x, y, 'ro', markersize=3)

ax.set_aspect('equal')
ax.set_title("Warren Truss Below Inclined Bridge Deck")
ax.set_xlabel("Span (m)")
ax.set_ylabel("Height (m)")
plt.grid(True)
plt.show()
```
